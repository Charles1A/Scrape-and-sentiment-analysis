```python
from bs4 import BeautifulSoup
import requests
import re

from datetime import datetime
import time

import pandas as pd

print('Pandas version number:', pd.__version__)
```

```python
def yelp_scraper():

  rlist = []

  for page in range(0, 150, 10):

    print(f"Scraping page {page}.")

    url = f'https://www.yelp.com/biz/nobu-malibu-malibu-4?start={page}&sort_by=date_desc'

    result = requests.get(url, allow_redirects=True)
    src = result.content
    soup = BeautifulSoup(src, 'lxml')

    entries = soup.find_all('li', class_="yelp-emotion-1jp2syp")

    for entry in entries:

      star_rating = entry.find('div', class_="yelp-emotion-9tnml4")
      date = entry.find('div', class_="yelp-emotion-19pbem2")
      commentary = entry.find('p', class_="comment__09f24__D0cxf yelp-emotion-19a3fsb")

      if star_rating != None and date != None and commentary != None:

        rlist.append(
        {
        'Date': date.find('span', class_="yelp-emotion-v293gj").text,
        'Star_rating': int(star_rating["aria-label"][0]),
        'Commentary': commentary.text
        }
        )

    time.sleep(3) # 5-second delay between calls to URL

  print("The number of reviews collected: ", len(rlist))
  return rlist
```

```python
# Run scraper function, place scraped info into dataframe:
yelp_df = pd.DataFrame(data=yelp_scraper())
```
